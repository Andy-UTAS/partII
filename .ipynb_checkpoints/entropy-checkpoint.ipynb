## Stretch that rubber band!

It would be lovely to have a calculation which tells us what will happen when we stretch a rubber band. Fortunately, we have the technology!

When a rubber band is stretched _adiabatically_ (that is, without heat exchange to the environment), its temperature changes due to the change in entropy. We can cook up a model which allows us to predict the temperature change. The temperature with length is related to the temperature dependence of the force via:

$$
\frac{dT}{dL} = \frac{T}{C_L} \left( \frac{\partial F}{\partial T} \right)_L
$$

where $T$ is the temperature, $L$ is the length of the rubber band, $C_L$ is the heat capacity at constant length, and $F$ is the tension force in the rubber.

We now assume the functional form of the force follows can be modelled the ideal chain model:

$$
F(T, L) = n k_B T \cdot \frac{L - L_0}{L_0}
$$

Where $n$ is the effective number of polymer chains which make up the rubber, $k_B$ is Boltzmannâ€™s constant, and $L_0$ is the natural length of the rubber band. 

If we take the partial derivative of $F$ with respect to $T$:

$$
\left( \frac{\partial F}{\partial T} \right)_L = n k_B \cdot \frac{L - L_0}{L_0}
$$

we find the governing equation

$$
\frac{dT}{dL} = \frac{n k_B}{C_L} \cdot \frac{L - L_0}{L_0} \cdot T .
$$

Now we are cooking with gas: we know how to solve differential equations, and this is a first-order, separable differential equation. But given we are doing computational work, we are going to solve this numerically! We shall also use the fact we know the exact solution to be

$$
T(L) = T_0 \cdot \exp\left( \frac{n k_B}{C_L} \cdot \frac{(L - L_0)^2}{2 L_0} \right)
$$

with $T_0$ the temperature when $L = L_0$.


```python
# === Constants and Parameters ===
mass = 0.010          # Rubber band mass in kg
c_rubber = 2000       # Specific heat capacity in J/kgÂ·K
C_L = mass * c_rubber # Total heat capacity (J/K)
n = 1e23              # Number of polymer segments

# === Stretching conditions ===
L0 = 0.10             # Initial length (m)
L1 = 0.20             # Final length (m)
T0 = 293              # Initial temperature (K)

# === Define stretch function ===
def f_L(L):
    return (L - L0) / L0

# === dT/dL for numerical solver ===
def dT_dL(L, T):
    return + (n * kb / C_L) * f_L(L) * T

# === Numerical integration using solve_ivp ===
solution = integrate.solve_ivp(
    fun=dT_dL,
    t_span=(L0, L1),
    y0=[T0],
    dense_output=True
)

# === Exact analytic solution ===
def T_exact(L):
    exponent = (n * kb / C_L) * ((L - L0)**2) / (2 * L0)
    return T0 * np.exp(exponent)

# === Evaluate solutions for plotting ===
L_vals = np.linspace(L0, L1, 1000)
T_num = solution.sol(L_vals)[0]
T_ex = T_exact(L_vals)

# === Plot both solutions ===
plt.plot(L_vals, T_num, label="Numerical (solve_ivp)", linewidth=2)
plt.plot(L_vals, T_ex, '--', label="Exact (analytic)", linewidth=2)
plt.xlabel("Length (m)")
plt.ylabel("Temperature (K)")
plt.title("Adiabatic Temperature Rise During Rubber Band Stretching")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Print final temperature difference ===
T_final_exact = T_ex[-1]
T_final_numeric = T_num[-1]
delta_T_exact = T_final_exact - T0
delta_T_diff = T_final_numeric - T_final_exact

print(f"Final temperature (exact): {T_final_exact:.4f} K")
print(f"Final temperature (numeric): {T_final_numeric:.4f} K")
print(f"Temperature increase (exact): {delta_T_exact:.4f} K")
print(f"Difference between solutions: {delta_T_diff:.4e} K")
```
